### Problem Statement and Goal of Track Popularity Analysis

The core purpose of this analysis is to answer the following question:

Are some tracks “viral hit events”, achieving unusually high popularity relative to the size of the artist’s existing audience, while most tracks perform at levels that are consistent with the artist’s follower base?

Unlike artist-level popularity, which reflects aggregate and often sustained momentum, track-level popularity captures short-term, event-driven attention. As a result, raw track popularity alone cannot distinguish true viral hits from tracks that perform well simply because they are released by already popular artists.

To address this, we use a residual-based framework that models track popularity as a function of artist follower count. This establishes an expected level of track popularity for artists of similar scale. The difference between a track’s actual popularity and this expected value, referred to as the track residual, captures whether a specific track is significantly overperforming or underperforming relative to what would normally be expected for that artist.

By standardising these track-level residuals, we identify statistically rare positive deviations as viral hit events. This approach allows us to isolate genuine, track-specific virality from popularity driven by artist scale, providing a foundation for identifying one-hit wonders and distinguishing isolated viral tracks from sustained artist-level breakout success.

```python
idx = dataframe.groupby("artist_name")["track_popularity"].idxmax()
artist_data = dataframe.loc[idx].copy()
```

```python
artist_data = artist_data.drop(
    columns=[
        "track_id",
        "track_number",
        "explicit",
        "artist_genres",
        "album_id",
        "album_name",
        "album_release_date",
        "album_total_tracks",
        "album_type",
        "track_duration_min"
    ],
    errors="ignore"
)
artist_data
```

```python
# conerting attributes to numerical type
artist_data["artist_followers"] = pd.to_numeric(artist_data["artist_followers"], errors='coerce').fillna(0)
artist_data["artist_popularity"] = pd.to_numeric(artist_data["artist_popularity"], errors='coerce').fillna(0)
```

```python
from sklearn.linear_model import HuberRegressor


# Transform number of followers (artist size baseline for tracks)
artist_data["log_followers"] = np.log1p(artist_data["artist_followers"])

# Fit baseline (robust): expected track popularity given artist followers
X = artist_data[["log_followers"]].values
y = artist_data["track_popularity"].values
track_model = HuberRegressor().fit(X, y)

artist_data["track_pop_expected"] = track_model.predict(X)

# Residuals + robust z for track-level virality
artist_data["track_residual"] = artist_data["track_popularity"] - artist_data["track_pop_expected"]

med = artist_data["track_residual"].median()
mad = (artist_data["track_residual"] - med).abs().median()
robust_sigma = 1.4826 * mad if mad > 0 else artist_data["track_residual"].std(ddof=0)

artist_data["track_z"] = (artist_data["track_residual"] - med) / (robust_sigma if robust_sigma > 0 else 1.0)

# Quantile bands for artist followers (same concept, but applied to track rows)
q30 = artist_data["artist_followers"].quantile(0.30)
q85 = artist_data["artist_followers"].quantile(0.85)

low = artist_data["artist_followers"] <= q30
high = artist_data["artist_followers"] >= q85

# Track-level labels (this is "viral track relative to artist size")
artist_data["track_segment"] = "Normal"

artist_data.loc[low & (artist_data["track_z"] >= 2) & (artist_data["track_popularity"] >= 70), "track_segment"] = "Viral track (small artist)"
artist_data.loc[(~low) & (~high) & (artist_data["track_z"] >= 2) & (artist_data["track_popularity"] >= 70), "track_segment"] = "Viral track (mid artist)"
artist_data.loc[high & (artist_data["track_z"] >= 2) & (artist_data["track_popularity"] >= 70), "track_segment"] = "Viral track (big artist)"

# Optional: an "underperforming track" label for big artists
artist_data.loc[high & (artist_data["track_z"] <= -1), "track_segment"] = "Underperforming track (big artist)"
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.scatter(
    artist_data["log_followers"],
    artist_data["track_popularity"],
    alpha=0.3
)

# expected popularity line
order = np.argsort(artist_data["log_followers"])
plt.plot(
    artist_data["log_followers"].iloc[order],
    artist_data["track_pop_expected"].iloc[order]
)

plt.xlabel("log(1 + followers)")
plt.ylabel("track popularity")
plt.title("track Popularity vs Followers with Expected Trend")
plt.show()
```



This plot illustrates the relationship between track popularity and artist follower count on a log-transformed scale, along with the estimated expected trend. On average, tracks released by artists with larger follower bases achieve higher popularity, reflecting the strong influence of artist scale on track performance.

The fitted trend line represents the expected track popularity for an artist of a given size and serves as the baseline for evaluating individual tracks. Most tracks cluster around this line, indicating that track success is generally consistent with the artist’s existing audience.

Tracks that lie significantly above the expected trend are outperforming expectations and represent viral hit events, where a single track gains unusually high attention relative to the artist’s follower base. In contrast, tracks below the trend underperform relative to expectation. This plot provides the foundation for isolating true track-level going viral from popularity driven primarily by artist scale.


```python
plt.figure()
plt.hist(artist_data["track_residual"], bins=50)
plt.xlabel("Residual (actual − expected popularity)")
plt.ylabel("Count")
plt.title("Track Residual Popularity Distribution")
plt.show()
```



This histogram shows the distribution of track-level residual popularity, defined as the difference between a track’s actual popularity and the popularity expected given the artist’s follower base.

The distribution is centered close to zero, indicating that most tracks perform broadly in line with expectations for the size of the artist releasing them. This confirms that artist scale explains a substantial portion of track performance for the majority of releases.

The positive tail of the distribution is relatively thin, showing that large positive residuals are rare. These extreme right-tail values correspond to viral hit events, where individual tracks achieve unusually high popularity relative to the artist’s existing audience. Their sparsity highlights that true track-level going viral is an outlier event rather than a common occurrence.

The negative side of the distribution is wider and more pronounced, indicating that underperforming tracks are more common than viral successes. This reflects typical release dynamics, where many tracks fail to gain traction while only a small fraction break through. Overall, the residual distribution supports the interpretation of viral tracks as statistically rare deviations from expected performance.


```python
plt.figure()
plt.scatter(
    artist_data["log_followers"],
    artist_data["track_residual"],
    alpha=0.3
)

plt.axhline(0)
plt.xlabel("log(1 + followers)")
plt.ylabel("Track Residual popularity")
plt.title("Track Residuals vs Followers")
plt.show()
```

```python
viral = artist_data[(artist_data["track_z"] >= 2) & (artist_data["track_popularity"] >= 70)]

plt.figure()
plt.scatter(artist_data["log_followers"], artist_data["track_residual"], alpha=0.15)
plt.scatter(viral["log_followers"], viral["track_residual"])
plt.axhline(0)
plt.xlabel("log(1 + followers)")
plt.ylabel("track Residual popularity")
plt.title("Viral Tracks Highlighted")
plt.show()
```



This plot shows track-level residual popularity plotted against log-transformed artist follower count after accounting for the expected relationship between artist size and track popularity. Residuals are centered around zero across the full range of follower counts, indicating that the influence of artist scale has been effectively removed.

Tracks with large positive residuals appear as isolated points above zero, primarily among low to mid–follower artists. These represent viral hit events where a single track dramatically outperforms expectations for the artist’s audience size. Their sparsity reinforces the conclusion that track-level virality is rare rather than a common outcome.

Tracks from artists with large follower bases cluster tightly around zero residual, suggesting that most releases by established artists perform close to expectation. The pronounced negative tail, especially at higher follower counts, indicates that underperforming tracks are more common than viral successes. Overall, this plot confirms that residual-based analysis cleanly separates event-driven track virality from popularity driven by artist scale.
